# LSTM
Если вы читаете недавние статьи, вы увидите, что многие люди используют вариант ванильной RNN, называемый RNN с долговременной памятью (LSTM). Ванильные RNN могут быть сложными для обучения на длинных последовательностях из-за исчезающих и взрывающихся градиентов, вызванных повторным умножением матриц. LSTM решают эту проблему, заменяя простое правило обновления ванильной RNN механизмом блокировки следующим образом.


Как и в обычной RNN, на каждом временном шаге мы получаем ввод $x_t\in\mathbb{R}^D$ и предыдущее скрытое состояние $h_{t-1}\in\mathbb{R}^H$; LSTM также поддерживает $H$-размерное *cell state*, поэтому мы также получаем предыдущее состояние ячейки $c_{t-1}\in\mathbb{R}^H$. Обучаемые параметры LSTM представляют собой матрицу *input-to-hidden* $W_x\in\mathbb{R}^{4H\times D}$, матрицу *hidden-to-hidden* $W_h\in\mathbb{R}^{4H\times H}$ и *вектор смещения* $b\in\mathbb{R}^{4H}$.


На каждом временном шаге мы сначала вычисляем *вектор активации* $a\in\mathbb{R}^{4H}$ как $a=W_xx_t + W_hh_{t-1}+b$. Затем мы делим это на четыре вектора $a_i,a_f,a_o,a_g\in\mathbb{R}^H$, где $a_i$ состоит из первых $H$ элементов $a$, $a_f$ — это следующие $H $ элементов $a$ и т. д. Затем мы вычисляем *input gate* $g\in\mathbb{R}^H$, *forget gate* $f\in\mathbb{R}^H$, *output gate* $o\in\mathbb{R}^H$ и *block input* $g\in\mathbb{R}^H$ как

$$
i = \sigma(a_i) \hspace{2pc}
f = \sigma(a_f) \hspace{2pc}
o = \sigma(a_o) \hspace{2pc}
g = \tanh(a_g)
$$

где $\sigma$ — сигмовидная функция, а $\tanh$ — гиперболический тангенс, оба применяются поэлементно.

Наконец, мы вычисляем следующее состояние ячейки $c_t$ и следующее скрытое состояние $h_t$ как

$$
c_{t} = f\odot c_{t-1} + i\odot g \hspace{4pc}
h_t = o\odot\tanh(c_t)
$$

где $\odot$ — поэлементное произведение векторов.

В оставшейся части записной книжки мы реализуем правило обновления LSTM и применим его к задаче создания подписей к изображениям.

В коде предполагается, что данные хранятся пакетами, так что $X_t \in \mathbb{R}^{N\times D}$, и будут работать с *транспонированными* версиями параметров: $W_x \in \mathbb {R}^{D \times 4H}$, $W_h \in \mathbb{R}^{H\times 4H}$, так что активации $A \in \mathbb{R}^{N\times 4H}$ могут эффективно вычисляться как $A = X_t W_x + H_{t-1} W_h$

# INLINE QUESTION
Напомним, что в LSTM input gate $i$, forget gate $f$, и output gate $o$ являются выходами сигмовидной функции. Почему бы нам не использовать функцию активации ReLU вместо сигмоида для вычисления этих значений? Объясните.

**Ответ:**

Значения этих гейтов при использовании сигмовидной функции всегда находятся в диапазоне (0,1). Так, например, гейт $forget$ обучается тому, с какими коэффициентами из интервала (0,1) нужно использовать информацию (значения) о состоянии, хранимую в ячейке $c_t$.  Где значения близкие к 0 означают, что информация сохраненная в ячейку $c_t$ забывается, а близкие к 1 - что используется без изменения. Если использовать вместо сигмоиды функцию ReLU, то значения гейтов могли бы неограниченно возрастать [0, +inf) и иметь разный масштаб, что усложнило бы обучение модели и могло привести к "взрыву" при прямом и обратном проходах обработки длинных последовательностей. И тем более гейты перестали бы выполнять функцию "переключателя", контролирующего долю пропускания информации: скрытого состояния/входных данных. К тому же гейты участвуют при протекании градиентов как через ячейку хранимого состояния $c_t$, так и скрытого состояния $h_t$. Здесь же проблема в том что через ReLU (не PReLU) не протекают градиенты для входов которые < 0. Это усугубляет протекание градиентов при обработке длинных последовательностей. И в таком случае не учитывается влияние гейтов на градиенты по таким входам.
